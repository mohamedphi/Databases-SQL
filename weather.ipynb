## weather project

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from datetime import datetime
import time

url='https://eg.freemeteo.com/weather/cairo/current-weather/location/?gid=360630&language=english&country=egypt'
response=requests.get(url)
soup=BeautifulSoup(response.text,'html.parser')
soup
```

```python
connection=sqlite3.connect('weather.db')

for i in range(11):
    
    url='https://eg.freemeteo.com/weather/cairo/current-weather/location/?gid=360630&language=english&country=egypt'
    response=requests.get(url)
    soup=BeautifulSoup(response.text,'html.parser')
    
    temp=float(soup.find('div',attrs={'class':'temp'}).get_text().split('°')[0])
    wind=float(soup.find('div',attrs={'class':'wind'}).find('span',attrs={'class':'white'}).find('b').get_text().strip())
    humidity=float(soup.find('div',attrs={'class':'stats'}).get_text().split()[2][:-1])
    visiblity= float(soup.find('div',attrs={'class':'stats'}).get_text().split()[6][:-1])
    pressure=float(soup.find('div',attrs={'class':'stats'}).get_text().split()[-1][:-2])
    now=datetime.now()
    time_now=now.strftime('%d/%m/%Y %H:%M:%S')
    
    cursor=connection.execute('insert into weather (temp,wind,humidity,visiblity,pressure,time) values (?,?,?,?,?,?)',
                             [temp,wind,humidity,visiblity,pressure,time_now])
    connection.commit()
    
    time.sleep(3)
    
    print(i)
    
```
